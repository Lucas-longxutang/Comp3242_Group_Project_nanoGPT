```python
# ==========================================
# 01_prepare.ipynb
# Run this cell ONCE to setup the project workspace.
# ==========================================

import os
import requests
import pickle
import numpy as np

# 1. Build Directory Structure
directories = ['data', 'models', 'output']
for directory in directories:
    os.makedirs(directory, exist_ok=True)
    print(f"Directory ready: ./{directory}/")

# 2. Download Tiny Shakespeare
input_file_path = os.path.join('data', 'input.txt')
if not os.path.exists(input_file_path):
    print("Downloading Tiny Shakespeare dataset...")
    data_url = 'https://raw.githubusercontent.com/karpathy/char-rnn/master/data/tinyshakespeare/input.txt'
    with open(input_file_path, 'w', encoding='utf-8') as f:
        f.write(requests.get(data_url).text)
else:
    print("Dataset already downloaded.")

# 3. Read Data and Build Vocabulary
with open(input_file_path, 'r', encoding='utf-8') as f:
    data = f.read()

chars = sorted(list(set(data)))
vocab_size = len(chars)
print(f"Dataset length: {len(data):,} characters")
print(f"Vocabulary size: {vocab_size} unique characters")

stoi = {ch: i for i, ch in enumerate(chars)}
itos = {i: ch for i, ch in enumerate(chars)}

def encode(s):
    return [stoi[c] for c in s]

# 4. Split and Export to Binary
n = len(data)
train_data = data[:int(n * 0.9)]
val_data = data[int(n * 0.9):]

train_ids = np.array(encode(train_data), dtype=np.uint16)
val_ids = np.array(encode(val_data), dtype=np.uint16)

train_ids.tofile(os.path.join('data', 'train.bin'))
val_ids.tofile(os.path.join('data', 'val.bin'))
print(f"Saved data/train.bin ({len(train_ids):,} tokens)")
print(f"Saved data/val.bin ({len(val_ids):,} tokens)")

# 5. Save Metadata
meta = {
    'vocab_size': vocab_size,
    'itos': itos,
    'stoi': stoi,
}
with open(os.path.join('data', 'meta.pkl'), 'wb') as f:
    pickle.dump(meta, f)
print("Saved data/meta.pkl successfully. Workspace setup complete!")
```
